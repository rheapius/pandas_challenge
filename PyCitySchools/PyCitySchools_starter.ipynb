```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path


# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools_total= len(school_data_complete["school_name"].value_counts())
students_total=len(school_data_complete["student_name"])
total_budget= school_data["budget"].sum()
average_maths= school_data_complete["maths_score"].mean()
average_reading= school_data_complete["reading_score"].mean()
passing_maths= (len(school_data_complete.loc[(school_data_complete["maths_score"]>=50),:])/(students_total))*100
passing_reading=(len(school_data_complete.loc[(school_data_complete["reading_score"]>=50),:])/(students_total))*100
overall= (len(school_data_complete.loc[(school_data_complete["reading_score"]>=50) \
                              & (school_data_complete["maths_score"]>=50),:])/(students_total))*100


LGA_summary= pd.DataFrame({"Total Schools":[schools_total],"Total Students":[students_total],
                                "Total Budget":[total_budget], "Average Maths Score":[average_maths],
                                "Average Reading Score":[average_reading], "% Passing Maths": [passing_maths], 
                                "% Passing Reading": [passing_reading], "% Overall Passing": [overall] })
LGA_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary_data= pd.DataFrame({
    'School Type':school_data_complete.groupby('school_name')['type'].first(),
    'Total Students': school_data_complete.groupby('school_name')['student_name'].count(),
    'Total Budget': school_data.groupby('school_name')['budget'].first(),
    'Per Student Budget': school_data_complete.groupby('school_name')['budget'].first() \
                           /school_data_complete.groupby('school_name')['student_name'].count(),
    'Average Maths Score': school_data_complete.groupby('school_name')['maths_score'].mean(),
    'Average Reading Score': school_data_complete.groupby('school_name')['reading_score'].mean(),
    '% Passing Maths': (school_data_complete.loc[school_data_complete['maths_score'] >= 50]
                        .groupby('school_name')['student_name'].count() /
                       school_data_complete.groupby('school_name')['student_name'].count()) * 100,
    '% Passing Reading': (school_data_complete.loc[school_data_complete['reading_score'] >= 50]
                           .groupby('school_name')['student_name'].count() /
                          school_data_complete.groupby('school_name')['student_name'].count()) * 100,
    '% Overall Passing': (school_data_complete.loc[(school_data_complete['reading_score'] >= 50) &
                                                   (school_data_complete['maths_score'] >= 50)]
                           .groupby('school_name')['student_name'].count() /
                          school_data_complete.groupby('school_name')['student_name'].count()) * 100
})
school_summary_data                                   
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Define spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Add a new column 'Spending Ranges (Per Student)' to school_data_complete based on the spending bins
school_data_complete['Spending Range (Per Student)'] = pd.cut(
    school_data_complete['budget'] /school_data_complete['size'],
    bins=spending_bins,
    labels=spending_labels,
    include_lowest=True
)

# Group by 'Spending Range (Per Student)' and calculate average scores and passing percentages
performance_by_spending = (
    school_data_complete.groupby('Spending Range (Per Student)')
    .apply(lambda x: pd.Series({
        'Average Maths Score': x['maths_score'].mean(),
        'Average Reading Score': x['reading_score'].mean(),
        '% Passing Maths': (x['maths_score'] >= 50).mean() * 100,
        '% Passing Reading': (x['reading_score'] >= 50).mean() * 100,
        'Overall Passing Rate': ((x['maths_score'] >= 50) & (x['reading_score'] >= 50)).mean() * 100
    }))
)

# Display the DataFrame
performance_by_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Define size bins and labels
size_bins = [0, 1000, 2000, 5000]
size_labels = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']

# Create a new column 'School Size' based on the bins
school_data_complete['School Size'] = pd.cut(
    school_data_complete['size'],
    bins=size_bins,
    labels=size_labels,
    include_lowest=True
)

# Define a function to calculate passing percentages
def passing_percentage(scores):
    return (scores >= 50).mean() * 100

# Calculate passing percentages separately
math_passing_percentages = school_data_complete.groupby('School Size')['maths_score'].agg(passing_percentage)
reading_passing_percentages = school_data_complete.groupby('School Size')['reading_score'].agg(passing_percentage)

# Combine the results into a DataFrame
performance_by_size = pd.DataFrame({
    'Average Maths Score': school_data_complete.groupby('School Size')['maths_score'].mean(),
    'Average Reading Score': school_data_complete.groupby('School Size')['reading_score'].mean(),
    '% Passing Maths': math_passing_percentages,
    '% Passing Reading': reading_passing_percentages
})

# Calculate overall passing rate
performance_by_size['Overall Passing Rate'] = (
    performance_by_size['% Passing Maths'] + performance_by_size['% Passing Reading']
) / 2


performance_by_size
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
# Define a function to calculate passing percentages
def passing_percentage(scores):
    return (scores >= 50).mean() * 100

# Group by 'School Type' and calculate average scores and passing percentages
performance_by_type = (
    school_data_complete.groupby('type')
    .agg(
        Average_Maths_Score=('maths_score', 'mean'),
        Average_Reading_Score=('reading_score', 'mean'),
        Passing_Maths_Percentage=('maths_score', passing_percentage),
        Passing_Reading_Percentage=('reading_score', passing_percentage)
    )
)

# Calculate overall passing rate
performance_by_type['Overall Passing Rate'] = (
    performance_by_type['Passing_Maths_Percentage'] + performance_by_type['Passing_Reading_Percentage']
) / 2

performance_by_type
```
